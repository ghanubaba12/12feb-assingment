```python
An Exception in Python is an event that occurs during the execution of a program that disrupts the normal flow of the program's instructions. Exceptions are raised when a program encounters an error condition that it cannot handle.

The difference between Exceptions and Syntax errors is that Syntax errors are errors in the syntax of the code itself and occur when the code is written in a way that is not valid according to the rules of the Python language. Syntax errors are usually detected by the Python interpreter when the code is being parsed and can be easily fixed by correcting the syntax.

On the other hand, Exceptions are errors that occur during the execution of a program and can be handled using try-except blocks. Exceptions can be raised for a variety of reasons, such as division by zero, file not found, or invalid input from the user. The try-except blocks allow the program to handle these exceptions gracefully and recover from the error, rather than crashing.

In summary, Syntax errors are caused by incorrect code syntax, while Exceptions are errors that occur during the execution of the code.
```

```python
"""Q3. Which Python statements are used to catch and handle exceptions? Explain with an example."""
```

```python
In Python, the try and except statements are used to catch and handle exceptions. The try statement defines a block of code that is being monitored for exceptions. The except statement defines a block of code that will be executed if an exception is raised within the try block.

Here's an example that demonstrates how the try and except statements are used to handle an exception:
```

```python
try:
    x = 10
    y = 0
    z = x / y
    print(z)
except ZeroDivisionError:
    print("Error: Division by zero")
```

```python
In this example, the code within the try block tries to divide 10 by 0, which will raise a ZeroDivisionError exception. The except statement is used to catch this exception and print an error message indicating that a division by zero has occurred.

Note that the except statement can be used to handle multiple types of exceptions by specifying multiple except blocks, each with its own exception type. This allows the program to handle different types of exceptions in different ways, if needed.
```

```python
try:
    x = 10
    y = 0
    z = x / y
    print(z)
except ZeroDivisionError:
    print("Error: Division by zero")
except ValueError:
    print("Error: Invalid value")
```

```python
n this example, the code within the try block is still trying to divide 10 by 0, which will raise a ZeroDivisionError exception. The first except statement will catch this exception and print an error message indicating that a division by zero has occurred. The second except statement will catch a ValueError exception, but it will not be executed in this case since a ZeroDivisionError was raised.
```

```python
"""Custom Exceptions in Python are user-defined exception classes that can be used to handle specific error conditions that are not covered by the built-in exceptions. Custom Exceptions are created by creating a new class that inherits from the Exception class or one of its subclasses.

Custom Exceptions are useful for several reasons:"""

"""Improving error handling: Custom Exceptions allow you to define specific error conditions that can be easily handled in a consistent and predictable way.

Improving code readability: Custom Exceptions can be used to give descriptive names to specific error conditions, making it easier to understand what went wrong in the code.

Separating error handling logic: Custom Exceptions can be used to separate error handling logic from the rest of the code, making the code cleaner and easier to maintain.

Here's an example that demonstrates how to create a custom exception in Python:"""


class InvalidInputError(Exception):
    def __init__(self, message):
        self.message = message

def divide(x, y):
    if y == 0:
        raise InvalidInputError("Division by zero not allowed")
    return x / y

try:
    result = divide(10, 0)
    print(result)
except InvalidInputError as e:
    print("Error:", e.message)
"""In this example, we've defined a custom exception class called InvalidInputError that inherits from the Exception class. This custom exception class has a message attribute that can be used to store an error message.

In the divide function, we've added a check to raise the InvalidInputError exception if the denominator y is equal to 0. This will ensure that a division by zero is not allowed.

Finally, the try-except block is used to handle the InvalidInputError exception. If the exception is raised, the error message stored in the message attribute will be printed.""""



```

```python
class CustomException(Exception):
    def __init__(self, message):
        self.message = message

def divide(x, y):
    if y == 0:
        raise CustomException("Division by zero is not allowed.")
    return x / y

try:
    result = divide(10, 0)
    print(result)
except CustomException as e:
    print("Error:", e.message)
```

```python
In this example, we've created a custom exception class called CustomException that inherits from the built-in Exception class. The CustomException class has a message attribute that can be used to store an error message.

In the divide function, we've added a check to raise the CustomException if the denominator y is equal to 0. This will ensure that a division by zero is not allowed.

Finally, the try-except block is used to handle the CustomException. If the exception is raised, the error message stored in the message attribute will be printed.
```
